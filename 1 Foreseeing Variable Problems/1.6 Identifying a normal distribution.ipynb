```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

###### Create a variable, x, with 200 observations that are normally distributed

```python
np.random.seed(29)
x = np.random.randn(200)
```

###### Create a dataframe with the x variable:

```python
data = pd.DataFrame([x]).T
data.columns = ['X']
```

```python
data.head()
```

##### Make a histogram and a density plot of the variable distribution

```python
sns.distplot(data['X'], bins=30)
```

##### Create and display a Q-Q plot to assess a normal distribution:

```python
stats.probplot(data['X'], dist='norm', plot=plt)
plt.show()
```

- we determined whether a variable is normally distributed with a histogram and a Q-Q plot.
- To display the variable distribution as a histogram and density plot, we used seaborn's distplot() method. By setting the bins argument to 30, we created 30 contiguous intervals for the histogram.
- To create the Q-Q plot, we used stats.probplot() from SciPy, which generated a plot of the quantiles for our x variable in the y-axis versus the quantiles of a theoretical normal distribution, which we indicated by setting the dist argument to norm, in the x-axis.
- Since x was normally distributed, its quantiles followed the quantiles of the theoretical distribution, so that the dots of the variable values fell along the 45-degree line.

```python
boston = pd.read_csv('housing.csv', delim_whitespace=True, header=None)
boston.head()
```

```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
                'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']

boston.columns = column_names
```

```python
boston.head()
```

```python
sns.distplot(boston['RM'], bins=30)
```

```python
sns.distplot(boston['LSTAT'], bins=30)
```

```python
stats.probplot(boston['RM'], dist="norm", plot=plt)
plt.show()
```

Most of the observations of RM fall on the 45 degree line, which suggests that the distribution is approximately Gaussian, with some deviation towards the larger and smaller values of the variable.

```python
stats.probplot(boston['CRIM'], dist="norm", plot=plt)
plt.show()
```

CRIM does not follow a Gaussian distribution as most of its observations deviate from the 45 degree line in the Q-Q plot.
